```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import manifold
```

```python
df = pd.read_csv('../input/mnist_784.csv')
# df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
fig, axs = plt.subplots(1, 1, figsize=(13, 8))
sns.countplot(data=df, ax=axs, x='target')
plt.show();
```

```python
n = 10_000
X = df.drop('target', axis=1).values[:n]
y = df['target'].values[:n]

tsne = manifold.TSNE(n_components=2, random_state=42)
embeddings = tsne.fit_transform(X)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(13, 8))
sns.scatterplot(ax=axs, x=embeddings[:, 0], y=embeddings[:, 1], hue=y,
                palette='tab10', s=10)
plt.show();
```

```python
df_train = pd.read_csv('../input/train.csv')
# df_train.drop('Unnamed: 0', axis=1, inplace=True)
df_train
```

```python
df_test = pd.read_csv('../input/test.csv')
# df_test.drop('Unnamed: 0', axis=1, inplace=True)
df_test
```

```python
def count_target_percent(df, set_name):
    n = len(df)
    df_count_percent = pd.DataFrame(df.groupby(by=["target"]).count()['0']/n)
    df_count_percent["set"] = set_name
    df_count_percent.columns = ["percent", "set"]
    return df_count_percent
```

```python
df_count_percent = count_target_percent(df, "full")
df_train_count_percent = count_target_percent(df, "train")
df_test_count_percent = count_target_percent(df, "test")

df_counts = pd.concat([df_count_percent, df_train_count_percent,
                       df_test_count_percent])
df_counts["target"] = df_counts.index

fig, axs = plt.subplots(1, 1, figsize=(13, 8))
sns.barplot(data=df_counts, ax=axs, x='target', y="percent", hue="set")
plt.show();
```
